###### Import libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
'''This is our prediction from a model wich classifies between two classes cats aand dogs
If the probability of being dog greater that the probability of being cat the result will be dog and vice versa'''
predicted_proba = np.array([4.1542e-04, 9.9958e-01])

# These are our classes
label = ['Cat', 'Dog']

# The model input(Image)
im = np.array(Image.open('dog.PNG'))
```

###### Q1)

```python
# Create the figure container
fig = plt.figure()

'''Creating 2 subplots, one for the model input and other 
for bar chart for output probabilities'''
fig1 = fig.add_subplot(1,2,1)
fig2 = fig.add_subplot(1,2,2)

# Showing the input in the fig1 subplot
fig1.imshow(im)
# Hide any ticks related to the image subplot
fig1.axis('off')

# making bar chart for the output probabilities
fig2.barh(label, predicted_proba, 0.3)
# handel the space beween both subplots
plt.tight_layout()
plt.show()
```

```python
# defining the labels placed on x-axis
label = ['G1', 'G2', 'G3', 'G4', 'G5']

# defining set of values for each gender
men_values = np.array([20, 35, 30, 35, 27])
women_values = np.array([25, 32, 34, 20, 25])

# create bar chart using predefined labels and gender arrays
plt.bar(label ,men_values, 0.5,color = 'blue')
plt.bar(label ,women_values, 0.5, bottom=men_values, color='orange')

# set axes labels
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend(['Man', 'Women'])
plt.show()
```
