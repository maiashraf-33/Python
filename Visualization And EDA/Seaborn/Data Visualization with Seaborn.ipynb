### relplot()

```python
# Creating subplots with col and row.
# create one scatter plot for each level of the variable "study_time", arranged in columns.
# relplot() lets you create subplots in a single figure.
sns.relplot(x="absences", y="G3", 
            data=df3,
            kind="scatter",
            col='study_time')

# Show plot
plt.show()
```

```python
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x="absences", y="G3", 
            data=df3,
            kind="scatter", 
            row="study_time")

# Show plot
plt.show()
```

```python
# Create a scatter plot of G1 vs. G3
sns.relplot(x='G1', y='G3', data=df3, kind='scatter')

# Show plot
plt.show()
```

```python
# Adjust to add subplots based on school support
'''
    subplots based on whether the student received support from the school ("schoolsup"), 
    ordered so that "yes" comes before "no".
'''

sns.relplot(x="G1", y="G3", 
            data=df3,
            kind="scatter",
            col='schoolsup',
            col_order=['yes', 'no'])

# Show plot
plt.show()
```

```python
# Adjust further to add subplots based on family support
'''
    Add row subplots based on whether the student received support from the family ("famsup"),
    ordered so that "yes" comes before "no". This will result in subplots based on two factors.
'''

sns.relplot(x="G1", y="G3", 
            data=df3,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row="famsup",
            row_order=["yes", "no"])

# Show plot
plt.show()
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head(2)
```

```python
# Create scatter plot of horsepower vs. mpg
sns.relplot(x='horsepower', y='mpg',data=mpg, kind='scatter', size='cylinders')

# Show plot
plt.show()
```

```python
# Create scatter plot of horsepower vs. mpg
'''
    To make this plot easier to read, use hue to vary the color of the points by
    the number of cylinders in the car ("cylinders")
'''
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders", hue='cylinders')

# Show plot
plt.show()
```

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration', y='mpg', data=mpg ,kind= 'scatter', hue='origin', style='origin')

# Show plot
plt.show()
```

```python
# Create line plot
sns.relplot(x='model_year',y='mpg',data=mpg, kind='line')

# Show plot
plt.show()
```

```python
# Make the shaded area show the standard deviation
'''
    Change the plot so the shaded area shows the standard deviation instead of the confidence interval for the mean.
'''
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')

# Show plot
plt.show()
```

```python
# Create line plot of model year vs. horsepower, with turning off the confidence intervals on the plot

sns.relplot(x='model_year', y='horsepower',data=mpg, kind='line', ci=None)

# Show plot
plt.show()
```

```python
# Create different lines for each country of origin ("origin") that vary in both line style and color.
# Change to create subgroups for country of origin
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, 
            style='origin',
            hue='origin')

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",
            markers=True,
            dashes=False)

# Show plot
plt.show()
```

### catplot()
**- Used to create categorical plots**<br>
**- Same advantages of relplot()**<br>
**- Easily create subplots with col= and row=**

```python
# countplot() vs. catplot()
sns.catplot(x='Internet usage', data=df2, kind='count')
plt.xticks(rotation=20)
plt.show()
```

```python
# Changing the orientation
sns.catplot(y="Internet usage", data=df2,kind="count")
plt.yticks(rotation=45)
plt.show()
```

```python
# create subplots with col
sns.catplot(x='Finances', data=df2, kind='count', col='Spiders')
plt.show()
```

**Confidence intervals**
- Lines show 95% conidence intervals for the mean
- Show s uncertainty about our estimate
- Assumes our data is a random sample

```python
# Bar plots
# Displays mean of quantitative variable per category.

sns.catplot(x='Gender', y='Pets',data=df2, kind='bar')
plt.show()
```

```python
df2['Pets'][df2['Gender'] == 'female'].mean()
```

```python
# Turning off confidence intervals
sns.catplot(x='Gender', y='Pets',data=df2, kind='bar', ci=None)
plt.show()
```

```python
# Rearrange the categories
order=list(df3['study_time'].value_counts().index)[::-1]

sns.catplot(x="study_time", y="G3",
            data=df3,
            kind="bar", order=order)

# Show plot
plt.show()
```

**Box plot**
- Shows the distribution of quantitative data
- See median, spread, skewness, and outliers
- Facilitates comparisons between groups


```python
sns.catplot(x='Spiders', y='Age', data=df2, kind='box')
plt.show()
```

```python
# Change the order of categories
sns.catplot(x='Spiders', y='Age', data=df2, kind='box', order=[4.0, 1.0, 2.0])
plt.show()
```

```python
# Omitting the outliers using 'sym'
sns.catplot(x='Spiders', y='Age', data=df2, kind='box', sym='')
plt.show()
```

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x='internet',y='G3',data=df3, kind='box', hue='location', sym='')

# Show plot
plt.show()
```

**Changing the whiskers using `whis`**<br>
. By default, the whiskers extend to 1.5 * the interquartile range<br>
. Make them extend to 2.0 * IQR: whis=2.0<br>
. Show the 5th and 95th percentiles: whis=[5, 95]<br>
. Show min and max values: whis=[0, 100]

```python
sns.catplot(x='Spiders', y='Age', data=df2, kind='box', whis=[5, 95])
plt.show()
```

```python
# Change the code to set the whiskers to extend to the min and max values.
sns.catplot(x='Spiders', y='Age', data=df2, kind='box', whis=[0, 100])
plt.show()
```

###  Adding titles
- title for FacetGrid --> fig.suptitl<br>
- title for AxesSubplot --> set_title

```python
# FacetGrid
sns.set_palette('tab10')
sns.set_style('white')
b = sns.catplot(x='Spiders', y='Age', data=df2, kind='box', whis=[5, 95])
b.fig.suptitle('New title', y=1.01) # y --> distance between the title and the plot
b.set(xlabel='Spiders categories', ylabel='Age')

plt.show()
```

```python
#  AxesSubplot 
l = sns.lineplot(x='model_year', y='horsepower',data=mpg, ci=None)
l.set_title('Line Ploy', y=1.03)
l.set(xlabel='Car Model Year', ylabel='Average MPG')
plt.show()
```
