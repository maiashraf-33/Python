```python
import numpy as np 
import matplotlib.pyplot as plt

# Suppose we have Population of 1000 people ages
population = np.random.randint(20, 50, 100)
# Actual population mean
population.mean()
```

### Apply Law of large numbers

```python
def law_of_large_numbers(population):
    """Calculates the mean for each subpopulation and calls plot function
       to show the relationship between the sample size and the population mean
    """
    sample_means = np.cumsum(population)/range(1, len(population)+1)
    plot(sample_means, population)

def plot(sample_means, population):
    """Draws a graph for samples and mean for each sample"""
    
    plt.axhline(population.mean(), color='red')
    plt.plot(sample_means, '-')
    plt.legend(('Population mean', 'Sample mean'), loc='lower right')
    plt.xlabel('Sample size')
    plt.ylabel('Mean for sample')
```

```python
# As you can see, the larger the sample, the closer it gets to the average population
law_of_large_numbers(population)
```
