# Problem 2: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.

a) Exponential random variable with $\lambda = 0.5$

```python
population1 = np.random.exponential(2,10000)

def CLT_exp(sample_number, sample_size):
    sample_means = []
    for _ in range(sample_number):
        sample_means.append(np.random.choice(population1, sample_size).mean())
    sample_means
    plt.hist(sample_means, bins=30)
    plt.show()
    
CLT_exp(2000,3)
```

```python
CLT_exp(2000,10)
```

```python
CLT_exp(2000,50)
```

b) Binomial random variable with $n = 20$ and $p = 0.2$

```python
population2 = np.random.binomial(20, 0.2, 10000)

def CLT_binom(sample_number, sample_size):
    sample_means = []
    for _ in range(sample_number):
        sample_means.append(np.random.choice(population2, sample_size).mean())
    sample_means
    plt.hist(sample_means, bins=25)
    plt.show()
    
CLT_binom(1000, 50)
```

```python
CLT_binom(1000, 100)
```

```python
CLT_binom(1000, 1000)
```

c) Uniform random variable with $a = 2$ and $b = 8$

```python
from scipy.stats import uniform 
population3 = uniform.rvs(2, 8, 20000)

def CLT_unif(sample_number, sample_size):
    sample_means = []
    for _ in range(sample_number):
        sample_means.append(np.random.choice(population3, sample_size).mean())
    
    plt.hist(sample_means, bins=25)
    plt.show()
    
CLT_unif(500, 10)
```

```python
CLT_unif(1000, 100)
```

```python
CLT_unif(1000, 2000)
```
